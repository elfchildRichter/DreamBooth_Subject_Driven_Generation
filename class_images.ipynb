## Class Images

leverage Stable Diffusion model before fine-tuning to generate 220 class images

```python
class_images_dir = f'{PATH}/class-images'
os.makedirs(class_images_dir, exist_ok=True)

tf.keras.mixed_precision.set_global_policy("mixed_float16")
model = keras_cv.models.StableDiffusion(img_width=RESOLUTION, img_height=RESOLUTION, jit_compile=True)

class_prompt = "a photo of a cat"
generate_and_save_images(class_images_dir, model, class_prompt, 20)
```
